```python
# Problem 1: Unique Words
sentence = input("Enter sentence: ")
words = sentence.split()
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
# Problem 2: Highest Salary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_employee = max(employees, key=employees.get)
print(f"Highest Salary: {highest_employee} - {employees[highest_employee]}")
```

```python
# Problem 3: Max and Min
numbers = [45, 22, 89, 10, 66]

max_val = numbers[0]
min_val = numbers[0]

for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

```python
# Problem 4: Products Above 1000
prices = [450, 1200, 899, 1500, 300]

count = 0
for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

```python
# Problem 5: Attendance Percentage
attendance = ["P", "P", "A", "P", "P"]

present = attendance.count("P")
percentage = (present / len(attendance)) * 100

print("Attendance Percentage:", percentage)
```

```python
# Problem 6: Unique Phone Numbers
phone_numbers = [9876543210, 9123456789, 9876543210]

unique_numbers = set(phone_numbers)
print("Unique phone numbers:", unique_numbers)
```

```python
# Problem 7: Character Frequency
text = "python"
freq = {}

for char in text:
    freq[char] = freq.get(char, 0) + 1

print(freq)
```

```python
# Problem 8: List to Tuple
lst = [10, 20, 30]
tpl = tuple(lst)

print("Tuple:", tpl)
```

```python
# Problem 9: Check Key in Dictionary
employees = {"Ravi": 75000, "Anita": 68000}
key = "Ravi"

if key in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
# Problem 10: Average Marks
marks = [70, 80, 65, 75]

average = sum(marks) / len(marks)
print("Average Marks:", average)
```

```python
# 1. User Login Check

username = "admin"
password = "1234"

input_username = "admin"
input_password = "1234"

if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# 2. Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("\nPass / Fail Analysis")
print("Passed:", pass_count)
print("Failed:", fail_count)

```

```python

# 3. Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_names.append(name.strip().lower())

print("\nCleaned Names:")
print(cleaned_names)
```

```python
# 4. Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

print("\nMessage Length Analysis")

for message in messages:
    print("Message:", message)
    print("Length:", len(message))
    if len(message) > 10:
        print("This message is longer than 10 characters.")
    print()

```

```python

# 5. Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR entries:", error_count)
```

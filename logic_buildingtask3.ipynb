```python
# Problem 1: Employee Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
max_score = max(employees.values())

# Find all employees with highest score
top_performers = [name for name, score in employees.items() if score == max_score]

print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {max_score})")
```

```python
# Problem 2: Keyword Frequency

import string

query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))

# Count frequency
words = query.split()
freq = {}

for word in words:
    freq[word] = freq.get(word, 0) + 1

# Display words appearing more than once
result = {k: v for k, v in freq.items() if v > 1}
print(result)
```

```python
# Problem 3: Valid Sensor Readings

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for index, value in enumerate(sensor_readings):
    if value % 2 == 0:  # Even readings only
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Problem 4: Email Domain Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# Problem 5: Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

avg_sales = sum(sales) / len(sales)

print("Sales Spikes:")
for day, value in enumerate(sales, start=1):
    if value > avg_sales * 1.3:  # 30% above average
        print(f"Day {day}: {value}")
```

```python
# Problem 6: Duplicate User IDs

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

freq = {}

for uid in user_ids:
    freq[uid] = freq.get(uid, 0) + 1

# Print duplicates
for uid, count in freq.items():
    if count > 1:
        print(f"{uid} → {count} times")
```

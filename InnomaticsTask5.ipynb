```python
#Problem 1
def analyze_engagement(likes):
    total = 0
    for l in likes:
        total += l

    status = "Viral Post" if total >= 1000 else "Normal Engagement"

    print("Total Likes:", total)
    print("Post Status:", status)

analyze_engagement([200, 300, 250, 300])
```

```python
#Problem 2
def medicine_stock(stock):
    status = "Low Stock Alert" if stock < 10 else "Stock Sufficient"
    print("Medicine Stock:", stock)
    print("Status:", status)

medicine_stock(6)
```

```python
#Problem 3
def rainfall_checker(data, required=70):
    total = 0
    for r in data:
        total += r

    avg = total / len(data)
    status = "Adequate Rainfall" if avg >= required else "Inadequate Rainfall"

    print("Average Rainfall:", round(avg))
    print("Rainfall Status:", status)

rainfall_checker([60, 80, 75, 70])
```

```python
 #Problem 4: Duplicate Account Detection
def detect_duplicates(usernames):
    has_duplicates = len(usernames) != len(set(usernames))
    print("Duplicate Accounts Found:", "Yes" if has_duplicates else "No")

detect_duplicates(["user1", "user2", "user1"])
```

```python
 #Problem 6: Premium Crop Price Filter
def premium_crops(prices):
    premium = []
    for p in prices:
        if p > 2000:
            premium.append(p)

    print("Premium Crops:", premium)

premium_crops([1500, 2500, 1800, 3200])
```

```python
#Problem 7: Application Health Checker
def app_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)

app_health(7)
```

```python
#Problem 8: Banking Transaction Limit Checker
def transaction_checker(amount):
    status = "Approved" if amount <= 50000 else "Rejected"
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

transaction_checker(60000)
```

```python
#Problem 9: Attendance Eligibility System
def attendance_checker(attendance):
    present = 0
    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100
    status = "Eligible" if percentage >= 75 else "Not Eligible"

    print("Attendance Percentage:", round(percentage, 1))
    print("Exam Eligibility:", status)

attendance_checker(["P", "P", "A", "P", "P"])
```

```python
#Problem 10: Smart Electricity Bill Analyzer
def electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Total Bill:", bill)
    print("Usage Level:", usage)

electricity_bill(250)
```

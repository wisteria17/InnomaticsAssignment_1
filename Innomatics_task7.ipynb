```python
# simple class to control a smart light
class SmartLight:
    def __init__(self, name):
        self.name = name  # store light name

    def turn_on(self):
        print(self.name, "is ON")

    def turn_off(self):
        print(self.name, "is OFF")


# example usage
light1 = SmartLight("Bedroom Light")
light1.turn_on()
```

```python
# class to store employee id card info
class EmployeeCard:
    def __init__(self, name, emp_id, dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept

    def show_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.dept)


emp1 = EmployeeCard("Rahul", "EMP102", "IT")
emp1.show_card()
```

```python
# storing contact details using a class
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


c1 = Contact("Anita", 9876543210)
c1.display()
```

```python
# class for printing a price tag
class Product:
    def __init__(self, pname, price):
        self.pname = pname
        self.price = price

    def show_tag(self):
        print("Product:", self.pname)
        print("Price: ₹" + str(self.price))


p1 = Product("Headphones", 2499)
p1.show_tag()
```

```python
# class to store movie rating
class Movie:
    def __init__(self, title, rating):
        self.title = title
        self.rating = rating

    def show_movie(self):
        print("Movie:", self.title)
        print("Rating:", self.rating, "/ 5")


m1 = Movie("Inception", 4.8)
m1.show_movie()
```

```python
# class for delivery details
class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def show_details(self):
        print("Delivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)


d1 = Delivery("Suman", "Hyderabad")
d1.show_details()
```

```python
# Problem 1: Mobile Recharge Validation

def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        if amount < 50:
            print("Recharge must be at least ₹50. Try again.")
        elif amount not in valid_plans:
            print("Invalid plan. Valid plans:", valid_plans)
        else:
            print("Recharge successful!")
            break

# Call function
# validate_recharge()
```

```python

# Problem 2: Inventory Reorder Alert

def check_inventory(products):
    for item, stock in products.items():
        if stock < 15:
            print(f"{item}: Reorder Alert")
        else:
            print(f"{item}: Stock OK")

inventory = {
    "Rice": 10,
    "Wheat": 20,
    "Oil": 8,
    "Sugar": 18
}

check_inventory(inventory)
```

```python

# Problem 3: Student Result

def student_result(marks):
    total = 0
    for m in marks:
        total += m

    avg = total / len(marks)

    if avg >= 50:
        return "Pass"
    else:
        return "Fail"

marks_list = [60, 45, 55, 70]
print("Result:", student_result(marks_list))
```

```python

# Problem 4: Cab Fare Calculator

def calculate_fare(distance, peak):
    fare = 50 + (distance * 12)
    if peak:
        fare *= 1.25
    return fare

while True:
    try:
        km = float(input("Enter distance (km): "))
        peak_hour = input("Peak hour? (yes/no): ").lower()

        peak = True if peak_hour == "yes" else False
        print("Estimated Fare: ₹", calculate_fare(km, peak))

        retry = input("Calculate again? (yes/no): ").lower()
        if retry != "yes":
            break
    except:
        print("Invalid input. Try again.")
```

```python

# Problem 5: Attendance Eligibility

def check_attendance(attendance):
    present = 0
    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance_list = ["P", "P", "A", "P", "P", "A", "P"]
print("Attendance Status:", check_attendance(attendance_list))
```

```python

# Problem 6: Password Strength Checker

def check_password(password):
    if len(password) < 8:
        return "Weak Password"

    has_digit = False
    has_special = False
    special_chars = "@#$"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

pwd = input("Enter password: ")
print(check_password(pwd))
```

```python
# simple parking tracker using IN / OUT logs
def parking_system(max_slots, logs):
    current = 0  # track current vehicles inside

    for entry in logs:
        if entry == "IN":
            current += 1
        elif entry == "OUT":
            # avoid negative count
            if current > 0:
                current -= 1

    # final parking status
    print("Currently Parked Vehicles:", current)

    if current >= max_slots:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")

parking_system(50, ["IN","IN","IN","OUT","IN","IN","OUT"])
```

```python
# estimate delivery time based on distance + conditions
def get_delivery_time(km, traffic_level, weather_type):
    eta = km * 5  # base assumption: 5 mins per km

    # add delay for traffic
    if traffic_level == "High":
        eta += 15
    elif traffic_level == "Medium":
        eta += 8

    # weather impact
    if weather_type == "Rainy":
        eta += 10

    print("Estimated Delivery Time:", eta, "minutes")

get_delivery_time(8, "High", "Rainy")
```

```python
# calculate how full the theatre is
def check_seats(total, bookings):
    booked = len(bookings)  # number of booked seats
    percent = (booked / total) * 100

    print("Occupancy:", int(percent), "%")

    # decide show status
    if percent == 100:
        print("Show Status: Housefull")
    elif percent >= 70:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")

check_seats(200, [1]*150)
```

```python
# find average cpu load and classify
def cpu_status(values):
    avg = sum(values) / len(values)  # average cpu
    print("Average CPU Load:", int(avg), "%")

    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")

cpu_status([45,60,70,85,90])
```

```python
# check which classroom resources are overused
def classroom_check(data):
    over = []  # store overused items

    for item in data:
        if data[item] > 8:  # threshold assumed as 8 hrs
            over.append(item)

    # print results
    if len(over) > 0:
        print("Overused Resources:", ", ".join(over))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

classroom_check({"Projector":6,"AC":9,"Lights":4})
```

```python
# handle event registrations and waitlist
def register_event(limit, people):
    if people > limit:
        extra = people - limit  # overflow users
        print("Confirmed Registrations:", limit)
        print("Waitlisted Users:", extra)
        print("Registration Status: Closed")
    else:
        print("Confirmed Registrations:", people)
        print("Waitlisted Users: 0")
        print("Registration Status: Open")

register_event(100, 105)
```
